```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis

```python
df = pd.read_csv('/diabetes.csv')
```

```python
df.head()
```

```python
df.describe()
```

Diabetes Counts

```python
df['Outcome'].value_counts()
```

```python
df.groupby('Outcome').mean()
```

Separating Data and Labels

```python
# split dataframe by excluding last column (response variable)
X = df.drop(columns = 'Outcome', axis = 1)
Y = df['Outcome']
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
# fit and transform the data to fit a standard scale (unit variance)
standardized_data = scaler.fit_transform(X)
```

```python
X = standardized_data
print(X)
```

Train/Test Split

```python
# test data will be 20%, train data will be 80%
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

Training the Model

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the SVM classifier
classifier.fit(X_train, Y_train)
```

Model Evaluation

```python
X_train_pred = classifier.predict(X_train)
train_data_accuracy = accuracy_score(X_train_pred, Y_train)
```

```python
print('Accuracy Score (train data): ', train_data_accuracy)
```

```python
X_test_pred = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_pred, Y_test)
```

```python
print('Accuracy Score (test data): ', test_data_accuracy)
```

Prediction System

```python
# medical bio data for all columns to predict diabetes
input_data = np.asarray((4, 110, 92, 0, 0, 37.6, 0.191, 30))

reshaped_input_data = input_data.reshape(1,-1)

# standardized data
standardized_input_data = scaler.transform(reshaped_input_data)
print(standardized_input_data)

# predict label of diabetes or not
prediction = classifier.predict(standardized_input_data)
print(prediction)
```

```python
if prediction[0] == 0:
  print("This patient does not have diabetes.")
else:
  print("This patient has diabetes.")
```
